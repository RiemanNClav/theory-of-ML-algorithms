```python
#mas de una columna. 
def Reemplazo_Media(matriz):
    X = matriz
    sumas = []
    posiciones = []
    dicc = {}
    promedios = []

    for i in range(len(X[0])):
        suma = 0
        idx = 0
        posiciones.append([])
        for j in range(len(X)):
            if X[j][i]>0 or X[j][i]<0:
                suma = suma + X[j][i]
                idx += 1
            else:
                posiciones[i].append(j)
        
        sumas.append(suma)
        promedio = sumas[i]/idx
        promedios.append(promedio)
        dicc[promedios[i]] = posiciones[i]


    for z in range(len(posiciones)):
        for z1 in posiciones[z]:
            X[z1][z] = promedios[z]

    
    
    return X
    
```

```python
def Categoricas_a_Dummies(data,Categories):
    for category in Categories:
        cat_dummies = pd.get_dummies(data[category], prefix=category)
        data_new = data.join(cat_dummies)
        data_new = data_new.drop([category], axis = 1) 
        data = data_new
        
    return data
    
```
